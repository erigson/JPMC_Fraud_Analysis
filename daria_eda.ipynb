```python
#changing gender to 1 and 2 int
df['gender'] = df['gender'].replace({'M':1, 'F': 2})
df['gender'].unique()
df.head()
```

```python
#dropping street
df.drop('street', inplace=True, axis=1)
df.head()
```

```python
#changing dob object to age int
from datetime import date,datetime
pd.to_datetime(df['dob'])
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today() #change to trans_date_time
    return today.year - born.year - ((today.month,today.day) < (born.month,born.day))
  
df['Age'] = df['dob'].apply(age)
df.drop('dob', inplace=True, axis=1)
df.head()
```

```python
#creating new loaction feature from lat and long
#ask about margin of error for long and lat
df['location_matches'] = np.where((abs(df['lat'] - df['merch_lat']) < 0.5) & (abs(df['long'] - df['merch_long']) < 0.5), True, False)

#dropping the lat,long,merch lat, and merch long
#df.drop('lat', inplace=True, axis=1)
#df.drop('long', inplace=True, axis=1)
#df.drop('merch_long', inplace=True, axis=1)
```
